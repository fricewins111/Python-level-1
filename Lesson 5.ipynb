```python
# Задача-1:
# Напишите небольшую консольную утилиту,
# позволяющую работать с папками текущей директории.
# Утилита должна иметь меню выбора действия, в котором будут пункты:
# 1. Перейти в папку
# 2. Просмотреть содержимое текущей папки
# 3. Удалить папку
# 4. Создать папку
# При выборе пунктов 1, 3, 4 программа запрашивает название папки
# и выводит результат действия: "Успешно создано/удалено/перешел",
# "Невозможно создать/удалить/перейти"

# Для решения данной задачи используйте алгоритмы из задания easy,
# оформленные в виде соответствующих функций,
# и импортированные в данный файл из easy.py

import os



def start():
    greatings = input(''' Приветствую! Доступны сл команды:
    [1] Перейти в папку
    [2] Просмотреть содержимое текущей папки
    [3] Удалить папку
    [4] Создать папку
    
    Для выхода введите q
    -------------------
    Введите команду:
    ''')
    return greatings

def direct():
    greating = input(r'''
    Введите абсолютный путь куда нужно перейти:
    
    ''')
    retval = os.getcwd()
    try:
        
        directory = os.chdir(retval)
        print('Вы перешли в', retval)
    except:
        print("ОШИБКА!")
        

def mk():
    folder = input('Введите название папки: \n')
    os.mkdir(folder)
    
def del_mk():
    folder = input('Введите название папки: \n')
    os.rmdir(folder)
    
def see_all():
    data = [i for i in os.listdir()]
    for i in data:
        print(i)
        

    




if __name__ == '__main__':
    data = start()
    while data != 'q':
        direct() if data == '1' else \
        see_all() if data == '2' else \
        del_mk() if data == '3' else \
        mk() if data == '4' else \
        print(data, ' - Я не знаю такую команду повторите ввод')
        print('\n\n', '*'*50)
        data = start()
        
        

```
