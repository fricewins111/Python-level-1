## Задача-1: Дано произвольное целое число (число заранее неизвестно).
Вывести поочередно цифры исходного числа (порядок вывода цифр неважен).
* Подсказки:
* постарайтесь решить задачу с применением арифметики и цикла while;
* при желании решите задачу с применением цикла for.

```python
from random import randint  # берем модуль рандом, чтобы не знать какое число у нас есть
from time import sleep      # берем модуль времени и метод засыпания, для красивости)))

number = str(randint(0, 999999999)) # генерируем рандомное число и приводим его к строке (чтобы for работал)

for i in number:   # проходимся по каждому числу в строке
    print(i, end='')# выводим число в консоль
    sleep(0.5)  # красиво выводим число)))
print('\r') # переход на новую строку опять таки для красивости)
```

# Задача-3: Запросите у пользователя его возраст.
Если ему есть 18 лет, выведите: "Доступ разрешен",
иначе "Извините, пользование данным ресурсом только с 18 лет"

```python
years = int(input('Ваш возраст: '))

# В одну строчку
print("Доступ разрешен") if years >= 18 else print("Извините, пользование данным ресурсом только с 18 лет")

# В несколько строчек

if years >= 18:
    print("Доступ разрешен")
else:
    print("Извините, пользование данным ресурсом только с 18 лет")
```

## Задача-1: Дано произвольное целое число, вывести самую большую цифру этого числа.
Например, дается x = 58375.
Нужно вывести максимальную цифру в данном числе, т.е. 8.
Подразумевается, что мы не знаем это число заранее.
Число приходит в виде целого беззнакового.
* Подсказки:
* постарайтесь решить задачу с применением арифметики и цикла while;
* при желании и понимании решите задачу с применением цикла for.

```python
from random import randint
number, a = str(randint(0, 999999999)), 0 # генерируем случайное число

#длинная запись

for i in number:
    if a < int(i):
        a = int(i)
    else:
        pass
print(number, 'Самая большая цифра это -', a)

# более короткая и производительная запись
from random import randint

number = str(randint(0,999999999))
number_dict = [i for i in number]
print(number, 'Самая большая цифра это -', max(number_dict))
```

## Задача-3: Напишите программу, вычисляющую корни квадратного уравнения вида ax² + bx + c = 0.
* Коэффициенты уравнения вводятся пользователем.
* Для вычисления квадратного корня воспользуйтесь функцией sqrt() модуля math:
* import math
* math.sqrt(4) - вычисляет корень числа 4

```python
from math import sqrt    # импортируем модуль вычисляющий логорифм из библиотеки math

print('Решаем ax² + bx + c = 0')

a,b,c = float(input('Введите a: ')), float(input('Введите b: ')), float(input('Введите c: ')) #просим ввести числа
D = b ** 2 - 4 * a * c


if a != 0:                                # если a не равно 0 - то выполняем код
    x1 = (-b + sqrt(abs(D))) / (2 * a)    # вычисляем дискриминант, к сожалению модуль sqrt не работает с 
    x2 = (-b - sqrt(abs(D))) / (2 * a)    # отрицательными числами, поэтому приводим к абсолютному значению функ abs()
                                          # abs(-1) == abs(1)
    print('Нет корней!') if D < 0 else print('x1 =', x1, 'x2 =', x2) if x1 != x2 else print('x =', x1)
# логика - Если дискриминант == 0, то корней нет, иначе если x1 != x2 то показать 2 значения иначе одно
else:
    print('a не может быть равна 0')
```
